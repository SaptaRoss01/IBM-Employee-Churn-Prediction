**Logistic Regression With Logit Link**

```python
from sklearn.model_selection import train_test_split as tts
data1["Attrition"]=[1 if i=="Yes" else
                   0 if i=="No" else
                    i for i in data1["Attrition"]]
data1["Attrition"]
```

```python
data2=pd.get_dummies(data1,drop_first=True,dtype=int);data2
```

```python
y=data2["Attrition"]
x=data2.drop("Attrition",axis=1)
```

```python
x_train,x_test,y_train,y_test=tts(x,y,train_size=0.8,random_state=42)
```

```python
#VIF Calculation
from statsmodels.stats.outliers_influence import variance_inflation_factor as VIF
import statsmodels.api as sm
d3=sm.add_constant(x_train[num])
vif=pd.DataFrame()
vif['variable']=d3.columns
vif['VIF']=[VIF(d3.values,i) for i in range(len(d3.columns))]
vif.iloc[1:,:]
```

```python
sm.Logit(y_train,x_train).fit().summary()
```

```python
#BackWard Feature Selection
def s(y1,X1):
    sig=list(X1.columns)
    while True:
        X_c = sm.add_constant(X1[sig])
        model = sm.Logit(y1, X_c).fit(disp=0)  # Setting disp=0 to suppress convergence warnings
        p_values = model.pvalues[1:]  # Exclude the constant term
        feature_to_remove = p_values.idxmax()
        if p_values.max()<0.05:
            break
        sig.remove(feature_to_remove)
    return sig
    
sig1=s(y_train,x_train);len(sig1)
```

```python
model=sm.Logit(y_train,sm.add_constant(x_train[sig1])).fit()
model.summary()
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay,accuracy_score,precision_score,recall_score,f1_score

pred=model.predict();pred
tpr=[]
fpr=[]
for p in pred:
    y_pred=(pred>p).astype(int)
    tn, fp, fn, tp =confusion_matrix(y_train,y_pred).ravel()
    tpr.append(tp/(tp+fn))
    fpr.append(fp/(fp+tn))
d=pd.DataFrame({"Tpr":tpr,"Fpr":fpr}).set_index(pred);d["s"]=d["Tpr"]*(1-d["Fpr"]);d
threshold1=d["s"].idxmax()
```

```python
threshold1 #Threshold probablity
```

```python
#ROC Curve
plt.plot(sorted(fpr),sorted(tpr))
plt.plot([0,1],[0,1])
plt.show() 
```

```python
y_p=(pred>threshold1).astype(int);y_p
```

```python
disp=ConfusionMatrixDisplay(confusion_matrix(y_train,y_p))
disp.plot(cmap="magma")
plt.show()
```

```python
#Measuring Parameters in Training Set
accuracy=accuracy_score(y_train,y_p);print(accuracy)
precision=precision_score(y_train,y_p);print(precision)
recall=recall_score(y_train,y_p);print(recall)
f1 = f1_score(y_train, y_p);print(f1)
```

```python
pred_test=model.predict(sm.add_constant(x_test[sig1]));pred_test
```

```python
y_p_t=(pred_test>threshold1).astype(int);y_p_t
disp=ConfusionMatrixDisplay(confusion_matrix(y_test,y_p_t))
disp.plot(cmap="magma")
plt.show()
```

```python
#Measuring Parameters in Testing Data
accuracy_t=accuracy_score(y_test,y_p_t);print(accuracy_t)
precision_t=precision_score(y_test,y_p_t);print(precision_t)
recall_t=recall_score(y_test,y_p_t);print(recall_t)
f1_t = f1_score(y_test, y_p_t);print(f1_t)
```

```python
#AIC Deviance Calculations
y_p_t = (pred_test > threshold1).astype(int)
likelihood = np.sum(y_p_t * np.log(pred_test) + (1 - y_p_t) * np.log(1 - pred_test))
k = len(model.params)
aic_value = 2 * k - 2 * likelihood
deviance_value = -2 * likelihood

print("AIC:", aic_value)
print("Deviance:", deviance_value)
```

***Building a Classification Tree***

```python
from sklearn.tree import DecisionTreeClassifier #To build a classification tree
from sklearn.tree import plot_tree #to draw a Classification Tree
from sklearn.metrics import confusion_matrix #To Create a Confusion Matrix
```

```python
#Create a Decision tree
clf_dt=DecisionTreeClassifier(random_state=42)
clf_dt=clf_dt.fit(x_train_sm,y_train_sm)
```

```python
plt.figure(figsize=(15,7.5))
plot_tree(clf_dt,filled=True,rounded=True,class_names=['Will not Leave','Will Leave'],
         feature_names=x_sm.columns);
```

```python
#Confusion Matrix
y_pred = clf_dt.predict(x_test_sm)
cm = confusion_matrix(y_test_sm, y_pred)
class_labels = ['Will Not Leave', 'Will Leave']
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='magma', xticklabels=class_labels, yticklabels=class_labels)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
```

In the Confusion matrix, We can see that of the **210+40=250** people that will not leave,**201(80.4%)** were correctly classified. And of the **244** people that will leave,**202(82.7%)** were correctly classified. It May have Overfitted the training Data.So lets prune the tree

pruning a decision tree is all about finding the right value of alpha which controls how little or how much pruning happens.One Way to find suitable alpha is to plot the accuracy of the tree as a function of different values.We will do this for both training and testing data


```python
#Cost complexitity Pruning
path=clf_dt.cost_complexity_pruning_path(x_train_sm,y_train_sm) #Determine values of alpha
ccp_alphas =path.ccp_alphas #extract different values of alpha
ccp_alphas= ccp_alphas[:-1] #Exclude the maxm value of alpha because this alpha will remove all leaves,leaving us with only a root instead of a tree
clf_dts =[] #an array that will put decision trees into

for ccp_alpha in ccp_alphas:
    clf_dt=DecisionTreeClassifier(random_state=0,ccp_alpha=ccp_alpha)
    clf_dt.fit(x_train_sm,y_train_sm)
    clf_dts.append(clf_dt)
```

```python
#Accuarcy as a Function of alpha
train_scores=[clf_dt.score(x_train_sm,y_train_sm) for clf_dt in clf_dts]
test_scores=[clf_dt.score(x_test_sm,y_test_sm) for clf_dt in clf_dts]
fig,ax=plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("Accuracy")
ax.set_title("Accuracy Vs alpha for training and testing data ")
ax.plot(ccp_alphas,train_scores,marker="o",label="train",drawstyle="steps-post")
ax.plot(ccp_alphas,test_scores,marker="o",label="test",drawstyle="steps-post")
ax.legend()
plt.show()
```

```python
#Cross Validation for finding the best alpha'
from sklearn.model_selection import cross_val_score
clf_dt=DecisionTreeClassifier(random_state=42,ccp_alpha=0.0026)
scores=cross_val_score(clf_dt,x_train_sm,y_train_sm,cv=5)
df=pd.DataFrame(data={'tree': range(5),'accuracy':scores}) 
df.plot(x='tree',y='accuracy',marker='o',linestyle='--')
```

```python
# create an array to store the results of each fold during cross validiation
alpha_loop_values = []

## For each candidate value for alpha, we will run 5-fold cross validation.

## Then we will store the mean and standard deviation of the scores (the accuracy) for each call #

# to cross_val_score in alpha_loop_values...

for ccp_alpha in ccp_alphas: 
    clf_dt= DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha) 
    scores= cross_val_score(clf_dt, x_train_sm, y_train_sm, cv=5) 
    alpha_loop_values.append([ccp_alpha, np.mean(scores), np.std(scores)])

# Now we can draw a graph of the means and standard deviations of the scores

## for each candidate value for alpha

alpha_results = pd.DataFrame(alpha_loop_values,

columns=['alpha', 'mean_accuracy', 'std'])

alpha_results.plot(x='alpha',

 y='mean_accuracy',

 yerr='std',

marker='o',

linestyle='-')
```

**We Can find the exact value of alpha with:**

```python
alpha_results[(alpha_results['alpha']>0.0015)
             &(alpha_results['alpha']<0.0025) ]
```

```python
ideal_ccp_alpha=0.001822
ideal_ccp_alpha=float(ideal_ccp_alpha)
```

```python
#Final Classification Tree
clf_dt_pruned=DecisionTreeClassifier(random_state=42,ccp_alpha=ideal_ccp_alpha)
clf_dt_pruned=clf_dt_pruned.fit(x_train_sm,y_train_sm)
```
